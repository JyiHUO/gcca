```python
from sklearn.metrics import r2_score
def cal_r2_score(list_view, list_U):
    left = list_U[0].T.dot(list_view[0])  # (top_r, N)
    right = list_U[1].T.dot(list_view[1])

    return r2_score(left, right)
```
